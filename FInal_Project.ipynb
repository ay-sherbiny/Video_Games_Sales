```python
plt.figure(figsize=(10, 6))
platform_sales.plot(kind='bar', color=['#FF4500', '#1E90FF', '#32CD32', '#FFD700', '#8A2BE2'])
plt.title('Total Global Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales (millions)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
avg_sales = df.groupby('Platform')['Global_Sales'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 5))
avg_sales.plot(kind='barh', color='black')
plt.title('Average Global Sales per Platform')
plt.xlabel('Average Sales (millions)')
plt.ylabel('Platform')
plt.gca().invert_yaxis()
plt.show()
```
